## SQL Queries

### 1. NÃºmero de rutas por pais 

```python
query1 = """
    select t2.pais_aeropuerto, count(t1.cod_aeropuerto_ori) as num_rutas from
    (
        select cod_aeropuerto_ori
        from HC_RUTAS

    ) t1 join 
    (
        select cod_aeropuerto, pais_aeropuerto from DM_AEROPUERTOS AS t2
    ) t2 on cod_aeropuerto_ori=cod_aeropuerto
    group by pais_aeropuerto
    order by num_rutas desc, cod_aeropuerto_ori
    """
```

```python
%sql $query1
```

### 2. NÃºmero de rutas de cada paÃ­s, como paÃ­s de la aerolÃ­nea ordenadas de forma que salga el primero el paÃ­s con mÃ¡s rutas. Mostrar el nombre del paÃ­s.

```python
query2 = """
    select t2.pais_aerolinea, count(t1.cod_aerolinea) as num_rutas from
    (
        select cod_aerolinea from HC_RUTAS
    ) as t1 join 
    (
        select cod_aerolinea, nombre_aerolinea, pais_aerolinea from DM_AEROLINEAS AS t2
    ) as t2 on t2.COD_AEROLINEA=t1.cod_aerolinea
    group by pais_aerolinea 
    order by num_rutas desc, pais_aerolinea
"""
```

```python
%sql $query2
```

### 3. NÃºmero de rutas de cada equipamiento ordenadas de forma que salga el primero el equipamiento con mÃ¡s rutas. Mostrar el nombre del equipamiento. Mostrar solo los equipamientos con mÃ¡s de 2000 rutas.

```python
query3="""
select t2.desc_equipamiento, t1.num_rutas from 
(
    select cod_equipamiento, count(cod_equipamiento) as num_rutas 
    from CO_RUTAS_EQUIPAMIENTOS
    group by cod_equipamiento
) t1 join
(
    select cod_equipamiento, desc_equipamiento from DM_EQUIPAMIENTOS
) t2 on t2.cod_equipamiento=t1.cod_equipamiento
where num_rutas > 2000
order by num_rutas desc, desc_equipamiento
"""
```

```python
%sql $query3
```

### 4. NÃºmero de rutas por equipamiento para las rutas cuyo aeropuerto origen es Barajas (1229) ordenadas de forma que salga el primero el equipamiento con mÃ¡s rutas. Mostrar el nombre del equipamiento.

```python
query4 = """
select t3.desc_equipamiento, count(t2.cod_ruta) as num_rutas from
(
    select cod_ruta, cod_equipamiento
    from CO_RUTAS_EQUIPAMIENTOS
) t1 inner join
(
    select cod_ruta, cod_aeropuerto_ori from HC_RUTAS where cod_aeropuerto_ori = 1229
) t2 on t1.cod_ruta=t2.cod_ruta join
(
    select cod_equipamiento, desc_equipamiento from DM_EQUIPAMIENTOS
) t3 on t1.cod_equipamiento=t3.cod_equipamiento
group by desc_equipamiento
order by num_rutas desc
"""
```

```python
%sql $query4
```

### 5.  Los dos paÃ­ses entre los que hay mÃ¡s rutas.

```python
query5 = """
SELECT t2.pais_aeropuerto AS Origen, t3.pais_aeropuerto AS Destino, COUNT(*) AS num_rutas FROM
(
    SELECT cod_aeropuerto_ori,cod_aeropuerto_des FROM HC_RUTAS
) t1 
JOIN 
(
    SELECT cod_aeropuerto, pais_aeropuerto FROM DM_AEROPUERTOS
) t2 ON t1.cod_aeropuerto_ori=t2.cod_aeropuerto
JOIN
(
   SELECT cod_aeropuerto, pais_aeropuerto FROM DM_AEROPUERTOS
) t3 ON t1.cod_aeropuerto_des=t3.cod_aeropuerto 
WHERE t2.pais_aeropuerto != t3.pais_aeropuerto
GROUP BY Origen,Destino
ORDER BY num_rutas desc
LIMIT 1
"""
```
